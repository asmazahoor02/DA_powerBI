# **Basics Data Visualization using Python**

## **MatPlotlib**

```python
# Import datasets
import pandas as pd

# reading the database
Orders = pd.DataFrame(pd.read_csv("/content/Orders_Large.csv"))

Orders.head()
```

```python
# Line Chart

import matplotlib.pyplot as plt
import numpy as np

# define data values
x = np.array([1, 2, 3, 4])  # X-axis points
y = x**3  # Y-axis points

plt.plot(x, y)  # Plot the chart
plt.show()  # display
```

```python
# Bar chart with Quantity against totalamount

plt.bar(Orders['Quantity'], Orders['TotalAmount'], color = 'Green')
plt.title("Relation between quantity and total amount of sale")

# Setting the X and Y labels
plt.xlabel('Quantity')
plt.ylabel('totalamount')

# Adding the legends
plt.show()
```

```python
# Scatter plot with Quantity against TotalAmount

plt.scatter(Orders['Quantity'], Orders['TotalAmount'])

# Adding Title to the Plot
plt.title("Scatter Plot")

# Setting the X and Y labels
plt.xlabel('Quantity')
plt.ylabel('TotalAmount')

plt.show()
```

```python
# histogram of Quantity

plt.hist(Orders['Quantity'])
plt.title("Histogram")

# Adding the legends
plt.show()
```

## **Seaborn**

```python
# scatter plot in seaborn

import seaborn as sns

tips = sns.load_dataset("tips")

sns.scatterplot(x="total_bill", y="tip", hue='time', size="size", sizes=(50, 200), data=tips)
```

```python
# Line chart in seaborn
sns.lineplot(x="day", y="tip", data=tips)
```

```python
# Bar chart in Seaborn

# load pre defined dataset in seaborn
titanic = sns.load_dataset("titanic")


# customize the bar plot
sns.barplot(x="class", y="fare", hue="who", ci=None, palette="muted", data=titanic)


# add labels and title
plt.xlabel("Class")
plt.ylabel("Fare")
plt.title("Average Fare by Class and Gender on the Titanic")

# display the plot
plt.show()
```

```python
# Histogram in seaborn

# Load iris datasets
iris = sns.load_dataset("iris")

# customize the histogram
sns.histplot(data=iris, x="petal_length", bins=20, kde=True, color="purple")

# add labels and title
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Petal Lengths in Iris Flowers")

# display the plot
plt.show()
```

```python
# density Plot

sns.kdeplot(data=tips, x="total_bill", hue="time", fill=True, alpha=0.6, linewidth=1.5)

# Add a title and labels to the plot using Matplotlib
plt.title("Density Plot of Total Bill by Meal Time")
plt.xlabel("Total Bill ($)")
plt.ylabel("Density")

# Show the plot
plt.show()
```

```python
# Box Plot

sns.boxplot(x="day", y="total_bill", hue="time", data=tips, palette="Set3", linewidth=1.5, fliersize=4)

# add a title, xlabel, and ylabel to the plot using Matplotlib functions
plt.title("Box Plot of Total Bill by Day and Meal Time")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")


# display the plot
plt.show()
```

```python
# Import Necessary Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create dataframe of same datasets

Orders = pd.DataFrame(pd.read_csv("/content/Orders_Large.csv"))
Customers = pd.DataFrame(pd.read_csv("/content/Customers_Large.csv"))
Products = pd.DataFrame(pd.read_csv("/content/Products_Large.csv"))
Sales = pd.DataFrame(pd.read_csv("/content/Sales_Large.csv"))
```

## **1. Create a histogram to understand data distribution**

```python
# HIstogram for data distribution


Orders.plot(kind = 'hist', bins=10, edgecolor= "black", title ='Total Amount', figsize=(15,7))
plt.xlabel("Total Revenue")
plt.show()
```

```python
# If we specify columns in hoistogram

plt.figure(figsize=(10, 7))
plt.hist(Orders['TotalAmount'], bins=10, color='pink', edgecolor='black')
plt.title('Distribution of Total Amount', fontsize=14)
plt.xlabel('Total Amount', fontsize=12)
plt.show()
```

## 2. Use a scatter plot to showcase relationships between two numerical variables.

```python
# Two numerical values in Scatter plot

Orders.plot(kind='scatter', x = 'ProductID',y = 'Quantity' , color ='Green', title =' Revenue vs Taxes')
plt.xlabel('ProductID', fontsize=12)
plt.ylabel('Quantity ', fontsize=12)
plt.show()
```

## **3. Develop a heatmap to highlight correlations.**

```python
# Create Correlations

correlation_matrix = Sales[['Revenue', 'Taxes']].corr()
print(correlation_matrix)
```

```python
# Create the heatmap

plt.figure(figsize = (10,8))
sns.heatmap(correlation_matrix, cmap = 'viridis', vmin = -1, vmax = 1, center = 0, cbar=True, fmt='.2f')
plt.title('Correlation Heatmap', fontsize=14)
plt.show()
```
